```python
import pandas as pd
from datetime import datetime
```

### T√°ch time ra ng√†y th√°ng nƒÉm

#### √âp ki·ªÉu thu·ªôc t√≠nh time t·ª´ str th√†nh datetime

```python
date_list = []

for date in df.time:
    date_list.append(date[:10])
```

#### t·∫°o 1 list ch·ª©a c√°c gi·ªù ƒëƒÉng b√†i

```python
hour_list = []

for time in df.time:
    time_obj = datetime.strptime(time, '%m/%d/%Y %H:%M')
    time = time_obj.time()
    hour_list.append(time.hour)
```

```python
hour_list
```

### T·∫°o dataframe ch·ª©a ng√†y v√† dataframe ch∆∞as gi·ª´o ƒëƒÉng b√†i

```python
df_date = pd.DataFrame(date_list, columns = ["date"])
df_hour = pd.DataFrame(hour_list, columns = ["hour"])
df_hour_date = pd.concat([df_hour, df_date], axis = 1)
# b·ªè c·ªôt time trong dataframe c≈©
no_time_df = df.drop(['time'], axis = 1)
# n·ªëi dataframe ƒë√£ x√≥a time v·ªõi df_hour_date
new_df = pd.concat([no_time_df, df_hour_date], axis = 1)
```

```python
new_df
```

### T·∫°o 1 list ch·ª©a th·ª© ƒëƒÉng b√†i

```python
day_of_week_list = []

for date in new_df.date:
    specific_date = pd.to_datetime(date)
    day_of_week = specific_date.strftime('%A')
    day_of_week_list.append(day_of_week)
```

### Th√™m th·ª© v√†o new_df

```python
day_df = pd.DataFrame(day_of_week_list, columns = ["day"])
new_df_2 = pd.concat([new_df, day_df], axis = 1)
```
