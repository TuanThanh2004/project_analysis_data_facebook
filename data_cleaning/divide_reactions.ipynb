### ƒê·ªçc file Lp_select_column.csv

```python
import pandas as pd
import ast
```

```python
df = pd.read_csv('C:/Users/Admin/aim/project_analysis_data_facebook/data/Lp_select_column.csv')
```

```python
df
```

### T·∫°o 1 list ch·ª©a c√°c dictionary c·ªßa c√°c lo·∫°i reaction c·ªßa t·ª´ng post v√† l∆∞u v√†o 1 list 

```python
react_list = []

for react_str in df.reactions:
    react_dict = ast.literal_eval(react_str)
    react_list.append(react_dict)
```

### T·∫°o dataframe t·ª´ react_list

```python
df_reactions = pd.DataFrame(react_list)
df_reactions
```

### Thay th·∫ø c√°c gi√° tr·ªã NaN b·∫±ng 0

```python
df_reactions = df_reactions.fillna(0)
```

### B·ªè thu·ªôc t√≠nh reactions c≈©

```python
no_react_df = df.drop(['reactions'], axis = 1)
```

### Thay thu·ªôc t√≠nh reactions b·∫±ng dataframe df_reactions

```python
new_df = pd.concat([no_react_df, df_reactions], axis = 1)
```
