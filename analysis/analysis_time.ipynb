#### ƒê·ªçc file data_ready.csv

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('C:/Users/Admin/aim/project_analysis_data_facebook/data/data_ready.csv')
```

```python
df
```

#### Th√™m c·ªôt Index

```python
df = df.reset_index(inplace = False)
df["index"] = df["index"] + 1
```

```python
df
```

#### Th·ªùi gian hay ƒëƒÉng b√†i 

```python
hour_plt = df.hour.value_counts().plot(kind = "bar", ylabel = 'frequency')

for b in hour_plt.patches:
    hour_plt.annotate(b.get_height(), (b.get_x() + b.get_width() / 2, b.get_height()), ha = 'center', va = 'center', xytext = (0,5), textcoords = 'offset points')

plt.show()
```

#### Bi·∫øn ƒë·ªông th·ªùi gian ƒëƒÉng b√†i

```python
df.plot(kind = "line", x = "index", y = "hour", drawstyle='steps', figsize=(10,6));
```

#### Bi·ªÉu ƒë·ªì bi·∫øn ƒë·ªông th·ªùi gian theo t·ª´ng ng√†y

```python
df['date'] = pd.to_datetime(df['date'])
df['date'].value_counts().sort_index().plot(kind='bar', figsize=(10,6));
```

```python
df.date.value_counts().plot(kind = "line", drawstyle = "steps", grid = True, figsize = (15, 7));
```
