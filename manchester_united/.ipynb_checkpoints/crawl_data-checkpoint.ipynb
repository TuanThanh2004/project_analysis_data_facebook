```python
from facebook_scraper import *
import pandas as pd
import numpy as np
```

```python
FANPAGE_LINK ="manchesterunited"
FOLDER_PATH = "C:/Users/Admin/aim/project_analysis_data_facebook/manchester_united/data/"
COOKIE_PATH = "C:/Users/Admin/aim/project_analysis_data_facebook/manchester_united/data/www.facebook.com_cookies.txt"
Crawl_data = ['text', 'time', 'reaction_count', 'reactions', 'comments', 'shares']

PAGES_NUMBER = 150
```

```python
post_list = []
run = 0
for post in get_posts(FANPAGE_LINK,
                    options={"comments": True, "reactions": True, "allow_extra_requests": True},
                    extra_info=True, pages=PAGES_NUMBER, cookies=COOKIE_PATH):
    
    test = True
    for i in range(len(Crawl_data)):
        if post.get(Crawl_data[i]) == 0:
            test = False
            print('error data')
            continue
    if test == True:
        post_list.append({
                        'post_id' : post.get('post_it'),
                        'text' : post.get('text'), 
                        'time' : post.get('time'), 
                        'reaction_count' : post.get('reaction_count'),
                        'reactions' : post.get('reactions'), 
                        'comments' : post.get('comments'), 
                        'comments_full' : post.get('comments_full'),
                        'shares' : post.get('shares') })
        print('Good data')
```

```python
post_list
```

```python
# Initialize dataframe to scrape Facebook post
post_df_full = pd.DataFrame(columns=post_list[0].keys(), index=range(len(post_list)), data=post_list)

# To df
path=FOLDER_PATH + FANPAGE_LINK + ".csv"
post_df_full.to_csv(path, index=False)
print(path)
```
